```python
import json
import matplotlib.pyplot as plt
import numpy as np
import os

plt.style.use('default')
```

```python
def load_json_file(path):
    with open(path, 'r') as file:
        data = json.load(file)
    return data

def get_axis_data(data):
    "return x,y axis from json data that download from tensorboard"
    epoch_axis = list(map(lambda x: x[1], data))
    val_axis = list(map(lambda x: x[2], data))
    return epoch_axis, val_axis

```

```python


base_path = "/Users/netanelblumenfeld/Desktop/bgu/Msc/Radar/my paper/data/ds_val_acc/"  
files = sorted(os.listdir(base_path))
for file_name,i in zip(files,[1,2,4,8]):
    path = base_path + file_name
    data = load_json_file(path)
    epoch_axis, val_axis = get_axis_data(data)
    plt.plot(epoch_axis, val_axis, label=f"down sample - {i}")
plt.grid(True, color='gray', linestyle='--', linewidth=0.5)
# # Setting label and title colors to black and removing background
plt.xlabel('Epochs', color='black', labelpad=10)
plt.ylabel('Accuracy', color='black', labelpad=10)
# plt.title('Accuracy per Epoch', color='black', pad=20)
# Setting the tick colors
plt.xticks(color='black')
plt.yticks(color='black')
plt.legend()

plt.show()
```
